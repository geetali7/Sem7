```python
import numpy as np
import matplotlib.pyplot as plt

def objective_function(x):
    """The function we want to minimize: f(x) = (x+3)²"""
    return (x + 3) ** 2

def derivative(x):
    """Derivative of the objective function: f'(x) = 2(x+3)"""
    return 2 * (x + 3)

def gradient_descent(starting_point, learning_rate, num_iterations):
    """
    Performs gradient descent optimization

    Args:
        starting_point (float): Initial x value
        learning_rate (float): Step size for updates
        num_iterations (int): Number of iterations to perform

    Returns:
        list: History of x values
        list: History of y values
    """
    x = starting_point
    x_history = [x]
    y_history = [objective_function(x)]

    for i in range(num_iterations):
        # Calculate gradient (derivative)
        grad = derivative(x)

        # Update x using gradient descent formula
        x = x - learning_rate * grad

        # Store values for plotting
        x_history.append(x)
        y_history.append(objective_function(x))

    return x_history, y_history

# Set parameters
starting_point = 2
learning_rate = 0.1
num_iterations = 20

# Run gradient descent
x_history, y_history = gradient_descent(starting_point, learning_rate, num_iterations)

# Create points for plotting the function
x_points = np.linspace(-5, 5, 100)
y_points = [objective_function(x) for x in x_points]

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(x_points, y_points, 'b-', label='f(x) = (x+3)²')
plt.plot(x_history, y_history, 'ro-', label='Gradient Descent Path')
plt.grid(True)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Gradient Descent Optimization')
plt.legend()
plt.show()

# Print final results
print(f"Starting point: x = {starting_point}, f(x) = {y_history[0]:.4f}")
print(f"Final point: x = {x_history[-1]:.4f}, f(x) = {y_history[-1]:.4f}")
```
